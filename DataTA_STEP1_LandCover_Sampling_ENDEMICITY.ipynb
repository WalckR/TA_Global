```python
import ee # Import the Python API package
import geemap # The geemap Python package is built upon the ipyleaflet and folium packages and implements several methods for interacting with Earth Engine data layer
import pandas as pd
try:
    ee.Initialize(project='amibea') # Initialize the API
except Exception as e:
    ee.Authenticate() # Authenticate to the Earth Engine servers
    ee.Initialize(project='amibea') # Initialize the API
```

```python
# Source the CSV file to your GEE Asset
Data = ee.FeatureCollection('projects/amibea/assets/TA_Endemicity_GEEformated_2026')
print('Size of Sample Collection',Data.size().getInfo())
```

```python
# Create a new map
Map1 = geemap.Map(center=[0,0], zoom=2)
Map1.addLayer(Data)
Map1
```

```python
##############################################################################################################################################################################################################
################################################################################ Loading and sampling land cover ################################################################################################
##############################################################################################################################################################################################################

# Create a bounding rectangle for standardization
unboundedGeo = ee.Geometry.Rectangle([-180, -88, 180, 88], "EPSG:4326", False)

# Load the land cover dataset
lc = ee.ImageCollection("projects/sat-io/open-datasets/GLC-FCS30D/five-years-map").select('b3').mosaic().rename('lc') # GLC_FCS30D Global 30-meter Land Cover Change Dataset (1985-2022)
# Sampling covariates
Data_Sampled = lc.sampleRegions(collection= Data,scale= 30,tileScale = 16,geometries=False)
```

```python
# Export to your drive and dowload manually to your working directory
taskParams = {'driveFolder': 'Amibatec','fileFormat': 'CSV'}
task = ee.batch.Export.table.toDrive(Data_Sampled,str('TA_SpeciesEndemicityPts_GEEformated_Sampled'),**taskParams)
task.start()
```
