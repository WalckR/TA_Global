```python
##############################################################################################################################################################################################################
################################################################################ Loading predictive variables ################################################################################################
##############################################################################################################################################################################################################

# Data are standardized by subtracting the global mean and dividing by its global standard deviation. 

# Create a bounding rectangle for standardization
unboundedGeo = ee.Geometry.Rectangle([-180, -88, 180, 88], "EPSG:4326", False)

##########################################
WorldClim = ee.Image('WORLDCLIM/V1/BIO')
##########################################

temp = WorldClim.select('bio01') # Annual Mean Temperature
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio01 = temp.subtract(ee.Image(mean.getInfo().get('bio01'))).divide(ee.Image(std.getInfo().get('bio01')))

temp = WorldClim.select('bio02')# Mean Diurnal Range
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio02 = temp.subtract(ee.Image(mean.getInfo().get('bio02'))).divide(ee.Image(std.getInfo().get('bio02')))

temp = WorldClim.select('bio03')# Isothermality
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio03 = temp.subtract(ee.Image(mean.getInfo().get('bio03'))).divide(ee.Image(std.getInfo().get('bio03')))

temp = WorldClim.select('bio04')# Temperature Seasonality
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio04 = temp.subtract(ee.Image(mean.getInfo().get('bio04'))).divide(ee.Image(std.getInfo().get('bio04')))

temp = WorldClim.select('bio05')# Max Temperature of Warmest Month
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio05 = temp.subtract(ee.Image(mean.getInfo().get('bio05'))).divide(ee.Image(std.getInfo().get('bio05')))

temp = WorldClim.select('bio06')# Min Temperature of Coldest Month
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio06 = temp.subtract(ee.Image(mean.getInfo().get('bio06'))).divide(ee.Image(std.getInfo().get('bio06')))

temp = WorldClim.select('bio07')# Temperature Annual Range (BIO5-BIO6)
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio07 = temp.subtract(ee.Image(mean.getInfo().get('bio07'))).divide(ee.Image(std.getInfo().get('bio07')))

temp = WorldClim.select('bio08')# Mean Temperature of Wettest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio08 = temp.subtract(ee.Image(mean.getInfo().get('bio08'))).divide(ee.Image(std.getInfo().get('bio08')))

temp = WorldClim.select('bio09')# Mean Temperature of Driest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio09 = temp.subtract(ee.Image(mean.getInfo().get('bio09'))).divide(ee.Image(std.getInfo().get('bio09')))

temp = WorldClim.select('bio10')# Mean Temperature of Warmest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio10 = temp.subtract(ee.Image(mean.getInfo().get('bio10'))).divide(ee.Image(std.getInfo().get('bio10')))

temp = WorldClim.select('bio11')# Mean Temperature of Coldest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio11 = temp.subtract(ee.Image(mean.getInfo().get('bio11'))).divide(ee.Image(std.getInfo().get('bio11')))

temp = WorldClim.select('bio12')# Annual Precipitation
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio12 = temp.subtract(ee.Image(mean.getInfo().get('bio12'))).divide(ee.Image(std.getInfo().get('bio12')))

temp = WorldClim.select('bio13')# Precipitation of Wettest Month
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio13 = temp.subtract(ee.Image(mean.getInfo().get('bio13'))).divide(ee.Image(std.getInfo().get('bio13')))

temp = WorldClim.select('bio14')# Precipitation of Driest Month
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio14 = temp.subtract(ee.Image(mean.getInfo().get('bio14'))).divide(ee.Image(std.getInfo().get('bio14')))

temp = WorldClim.select('bio15')# Precipitation Seasonality (Coefficient of Variation)
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio15 = temp.subtract(ee.Image(mean.getInfo().get('bio15'))).divide(ee.Image(std.getInfo().get('bio15')))

temp = WorldClim.select('bio16')# Precipitation of Wettest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio16 = temp.subtract(ee.Image(mean.getInfo().get('bio16'))).divide(ee.Image(std.getInfo().get('bio16')))

temp = WorldClim.select('bio17')# Precipitation of Driest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio17 = temp.subtract(ee.Image(mean.getInfo().get('bio17'))).divide(ee.Image(std.getInfo().get('bio17')))

temp = WorldClim.select('bio18')# Precipitation of Warmest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio18 = temp.subtract(ee.Image(mean.getInfo().get('bio18'))).divide(ee.Image(std.getInfo().get('bio18')))

temp = WorldClim.select('bio19')# Precipitation of Coldest Quarter
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
bio19 = temp.subtract(ee.Image(mean.getInfo().get('bio19'))).divide(ee.Image(std.getInfo().get('bio19')))

###################################################################################
TerraClim = ee.ImageCollection('IDAHO_EPSCOR/TERRACLIMATE')
###################################################################################

temp = TerraClim.select('aet').mean().rename('aet')  # Evapotranspiration
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
aet = temp.subtract(ee.Image(mean.getInfo().get('aet'))).divide(ee.Image(std.getInfo().get('aet')))

temp = TerraClim.select('def').mean().rename('def_') # Climate water deficit
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
def_ = temp.subtract(ee.Image(mean.getInfo().get('def_'))).divide(ee.Image(std.getInfo().get('def_')))

temp = TerraClim.select('pdsi').mean().rename('pdsi') # Palmer Drought Severity Index
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
pdsi = temp.subtract(ee.Image(mean.getInfo().get('pdsi'))).divide(ee.Image(std.getInfo().get('pdsi')))

temp = TerraClim.select('pet').mean().rename('pet') # Reference evapotranspiration
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
pet = temp.subtract(ee.Image(mean.getInfo().get('pet'))).divide(ee.Image(std.getInfo().get('pet')))

temp = TerraClim.select('pr').mean().rename('pr') # Precipitation accumulation
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
pr = temp.subtract(ee.Image(mean.getInfo().get('pr'))).divide(ee.Image(std.getInfo().get('pr')))

temp = TerraClim.select('soil').mean().rename('soil') # Soil moisture
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
soil = temp.subtract(ee.Image(mean.getInfo().get('soil'))).divide(ee.Image(std.getInfo().get('soil')))

temp = TerraClim.select('tmmn').mean().rename('tmmn') # Minimum temperature
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
tmmn = temp.subtract(ee.Image(mean.getInfo().get('tmmn'))).divide(ee.Image(std.getInfo().get('tmmn')))

temp = TerraClim.select('tmmx').mean().rename('tmmx') # Maximum temperature
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
tmmx = temp.subtract(ee.Image(mean.getInfo().get('tmmx'))).divide(ee.Image(std.getInfo().get('tmmx')))

temp = TerraClim.select('vap').mean().rename('vap') # Vapor pressure
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
vap = temp.subtract(ee.Image(mean.getInfo().get('vap'))).divide(ee.Image(std.getInfo().get('vap')))

temp = TerraClim.select('vpd').mean().rename('vpd') # Vapor pressure deficit
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
vpd = temp.subtract(ee.Image(mean.getInfo().get('vpd'))).divide(ee.Image(std.getInfo().get('vpd')))

temp = TerraClim.select('vs').mean().rename('vs') # Wind-speed at 10m
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
vs = temp.subtract(ee.Image(mean.getInfo().get('vs'))).divide(ee.Image(std.getInfo().get('vs')))

###################################################################################
MOD10A1 = ee.ImageCollection('MODIS/061/MOD10A1')
###################################################################################

temp = MOD10A1.select('NDSI_Snow_Cover').mean().rename('snowCover') # Snow cover
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
snowCover = temp.subtract(ee.Image(mean.getInfo().get('snowCover'))).divide(ee.Image(std.getInfo().get('snowCover'))).rename('snowCover')

###################################################################################
# Mean soil characteristics between 0-30 cm in depth
###################################################################################

temp = ee.Image("OpenLandMap/SOL/SOL_BULKDENS-FINEEARTH_USDA-4A1H_M/v02").select(['b0','b10','b30']).reduce(ee.Reducer.mean()).unmask().rename('bulk') # Soil bulk density
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale= 250)
bulk = temp.subtract(ee.Image(mean.getInfo().get('bulk'))).divide(ee.Image(std.getInfo().get('bulk')))

temp = ee.Image("OpenLandMap/SOL/SOL_CLAY-WFRACTION_USDA-3A1A1A_M/v02").select(['b0','b10','b30']).reduce(ee.Reducer.mean()).unmask().rename('clay')# Soil bulk density
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
clay = temp.subtract(ee.Image(mean.getInfo().get('clay'))).divide(ee.Image(std.getInfo().get('clay')))

temp = ee.Image("OpenLandMap/SOL/SOL_ORGANIC-CARBON_USDA-6A1C_M/v02").select(['b0','b10','b30']).reduce(ee.Reducer.mean()).rename('soc')# Soil organic carbon content
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
soc = temp.subtract(ee.Image(mean.getInfo().get('soc'))).divide(ee.Image(std.getInfo().get('soc')))

temp = ee.Image("OpenLandMap/SOL/SOL_PH-H2O_USDA-4C1A2A_M/v02").select(['b0','b10','b30']).reduce(ee.Reducer.mean()).rename('ph')# Soil pH in H2O
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
ph = temp.subtract(ee.Image(mean.getInfo().get('ph'))).divide(ee.Image(std.getInfo().get('ph')))

temp = ee.Image("OpenLandMap/SOL/SOL_SAND-WFRACTION_USDA-3A1A1A_M/v02").select(['b0','b10','b30']).reduce(ee.Reducer.mean()).rename('sand')# Sand content
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
sand = temp.subtract(ee.Image(mean.getInfo().get('sand'))).divide(ee.Image(std.getInfo().get('sand')))

temp = ee.Image("OpenLandMap/SOL/SOL_WATERCONTENT-33KPA_USDA-4B1C_M/v01").select(['b0','b10','b30']).reduce(ee.Reducer.mean()).rename('water')# Soil water content
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
water = temp.subtract(ee.Image(mean.getInfo().get('water'))).divide(ee.Image(std.getInfo().get('water')))

####################################################################################
# Vegetation variables
###################################################################################

MOD44B = ee.ImageCollection('MODIS/061/MOD44B')
temp = MOD44B.select('Percent_Tree_Cover').mean().rename('treeCover') # Percent tree cover
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
treeCover = temp.subtract(ee.Image(mean.getInfo().get('treeCover'))).divide(ee.Image(std.getInfo().get('treeCover')))

temp = MOD44B.select('Percent_NonTree_Vegetation').mean().rename('nontreeCover') # Percent non tree cover
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
nontreeCover = temp.subtract(ee.Image(mean.getInfo().get('nontreeCover'))).divide(ee.Image(std.getInfo().get('nontreeCover')))

temp = MOD44B.select('Percent_NonVegetated').mean().rename('nonvegetated') # Percent non vegetated
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
nonvegetated = temp.subtract(ee.Image(mean.getInfo().get('nonvegetated'))).divide(ee.Image(std.getInfo().get('nonvegetated')))

MOD13Q1 = ee.ImageCollection('MODIS/061/MOD13Q1')
temp = MOD13Q1.select('NDVI').mean().rename('ndvi')# Normalized Difference Vegetation Index
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
ndvi = temp.subtract(ee.Image(mean.getInfo().get('ndvi'))).divide(ee.Image(std.getInfo().get('ndvi')))

temp = MOD13Q1.select('EVI').mean().rename('evi')# Enhanced Vegetation Index
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
evi = temp.subtract(ee.Image(mean.getInfo().get('evi'))).divide(ee.Image(std.getInfo().get('evi')))

MOD17A2H = ee.ImageCollection('MODIS/061/MOD17A2H')
temp = MOD17A2H.select('Gpp').mean().rename('gpp')# Gross primary production
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
gpp = temp.subtract(ee.Image(mean.getInfo().get('gpp'))).divide(ee.Image(std.getInfo().get('gpp')))

temp = MOD17A2H.select('PsnNet').mean().rename('npp')# Net photosynthesis
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
npp = temp.subtract(ee.Image(mean.getInfo().get('npp'))).divide(ee.Image(std.getInfo().get('npp')))

MOD17A2H = ee.ImageCollection('MODIS/061/MCD15A3H')
temp = MOD17A2H.select('Lai').mean().rename('lai')# leaf area index
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True,scale = 250)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True,scale = 250)
lai = temp.subtract(ee.Image(mean.getInfo().get('lai'))).divide(ee.Image(std.getInfo().get('lai')))

biomes = ee.Image('OpenLandMap/PNV/PNV_BIOME-TYPE_BIOME00K_C/v01').rename('biomes')# Potential distribution of biomes

ecoregions = ee.FeatureCollection("RESOLVE/ECOREGIONS/2017").reduceToImage(['ECO_ID'], ee.Reducer.first()).rename('ecoregions'); # The RESOLVE Ecoregions dataset, updated in 2017

##################################################################################
# Topographic variables
###################################################################################

temp = ee.Image('USGS/GMTED2010_FULL').select('mea').rename('height')# Elevation
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
height = temp.subtract(ee.Image(mean.getInfo().get('height'))).divide(ee.Image(std.getInfo().get('height')))

lon = ee.Image.pixelCoordinates('EPSG:4326').select('x').rename('lon') # 
lat = ee.Image.pixelCoordinates('EPSG:4326').select('y').rename('lat') # 

#################################################################################
# Human variables
###################################################################################

temp = ee.ImageCollection("CIESIN/GPWv411/GPW_Population_Density").first().select('population_density').rename('pop')# The estimated number of persons per square kilometer
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
pop = temp.subtract(ee.Image(mean.getInfo().get('pop'))).divide(ee.Image(std.getInfo().get('pop')))

temp = ee.ImageCollection('CSP/HM/GlobalHumanModification').first().select('gHM').rename('gHM')# The Global Human Modification
mean = temp.reduceRegion(reducer= "mean",geometry= unboundedGeo,bestEffort= True)
std = temp.reduceRegion(reducer= "stdDev",geometry= unboundedGeo,bestEffort= True)
gHM = temp.subtract(ee.Image(mean.getInfo().get('gHM'))).divide(ee.Image(std.getInfo().get('gHM')))

###################################################################################
# Stacking environmental variable in a multiband image
###################################################################################

compositeOfInterest = biomes.addBands(ecoregions).addBands(bio01).addBands(bio02).addBands(bio03).addBands(bio04).addBands(bio05).addBands(bio06).addBands(bio07).addBands(bio08).addBands(bio09).addBands(bio10).addBands(bio11).addBands(bio12).addBands(bio13).addBands(bio14).addBands(bio15).addBands(bio16).addBands(bio17).addBands(bio18).addBands(bio19).addBands(aet).addBands(def_).addBands(pdsi).addBands(pet).addBands(pr).addBands(soil).addBands(tmmn).addBands(tmmx).addBands(vap).addBands(vpd).addBands(vs).addBands(snowCover).addBands(bulk).addBands(clay).addBands(soc).addBands(ph).addBands(sand).addBands(water).addBands(treeCover).addBands(nontreeCover).addBands(nonvegetated).addBands(ndvi).addBands(evi).addBands(gpp).addBands(npp).addBands(lai).addBands(height).addBands(lat).addBands(lon).addBands(pop).addBands(gHM)
```
